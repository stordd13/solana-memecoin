```python
# Cell 8: Match tokens between JSON and parquet files
print("\n\n=== Matching tokens between JSON and parquet files ===")

# Create a mapping of token addresses to metadata
token_map = {token['address']: token for token in metadata['tokens']}

# Get all parquet file addresses
parquet_token_info = []
for pf in parquet_files[:10]:  # Check first 10
    parts = pf.name.split('_')
    token_name = parts[0]
    token_address = parts[-1].replace('.parquet', '')
    
    # Check if this address exists in our JSON metadata
    token_data = token_map.get(token_address, None)
    
    if token_data:
        parquet_token_info.append({
            'file': pf.name,
            'name': token_name,
            'address': token_address,
            'has_metadata': True,
            'symbol': token_data['symbol'],
            'liquidity': token_data['liquidity'],
            'mc': token_data['mc'],
            'v24hUSD': token_data['v24hUSD']
        })
    else:
        parquet_token_info.append({
            'file': pf.name,
            'name': token_name,
            'address': token_address,
            'has_metadata': False
        })

# Display results
print(f"Matched {sum(1 for t in parquet_token_info if t['has_metadata'])} out of {len(parquet_token_info)} tokens\n")

for info in parquet_token_info[:5]:
    print(f"Token: {info['name']}")
    if info['has_metadata']:
        print(f"  Symbol: {info['symbol']}")
        print(f"  Liquidity: ${info['liquidity']:,.2f}")
        print(f"  Market Cap: ${info['mc']:,.2f}")
        print(f"  24h Volume: ${info['v24hUSD']:,.2f}")
    else:
        print(f"  No metadata found in JSON")
    print()
```

```python
# Cell 11: Data Quality Analysis
print("\n\n=== Data Quality Analysis ===")

# Sample 10 random parquet files for quality check
sample_size = min(10, len(parquet_files))
sampled_files = random.sample(parquet_files, sample_size)

quality_reports = []

for pf in sampled_files:
    df = pd.read_parquet(pf)
    token_name = pf.name.split('_')[0]
    
    # Basic info
    total_rows = len(df)
    date_range = df['datetime'].max() - df['datetime'].min()
    expected_minutes = int(date_range.total_seconds() / 60) + 1
    
    # Check for duplicates
    duplicate_times = df['datetime'].duplicated().sum()
    
    # Check for missing values
    missing_values = df.isnull().sum().sum()
    
    # Check time gaps
    df_sorted = df.sort_values('datetime')
    time_diffs = df_sorted['datetime'].diff()
    
    # Expected 1-minute intervals
    gaps = time_diffs[time_diffs > pd.Timedelta(minutes=1)]
    
    # Check for zero or negative prices
    zero_prices = (df['price'] <= 0).sum()
    
    # Check price outliers (sudden jumps)
    price_changes = df_sorted['price'].pct_change()
    extreme_changes = price_changes[price_changes.abs() > 10]  # >1000% change
    
    quality_reports.append({
        'token': token_name,
        'total_rows': total_rows,
        'expected_rows': expected_minutes,
        'missing_rows': expected_minutes - total_rows,
        'duplicate_times': duplicate_times,
        'missing_values': missing_values,
        'time_gaps': len(gaps),
        'zero_prices': zero_prices,
        'extreme_changes': len(extreme_changes),
        'date_range_hours': date_range.total_seconds() / 3600
    })

# Display summary
print("Data Quality Summary for 10 sampled tokens:\n")
print(f"{'Token':<15} {'Rows':<8} {'Expected':<8} {'Missing':<8} {'Gaps':<6} {'Dupes':<6} {'Zero$':<6}")

for report in quality_reports:
    print(f"{report['token']:<15} {report['total_rows']:<8} {report['expected_rows']:<8} "
          f"{report['missing_rows']:<8} {report['time_gaps']:<6} {report['duplicate_times']:<6} "
          f"{report['zero_prices']:<6} {report['extreme_changes']:<8}")

# Detailed analysis of gaps for first token with issues
print("\n\nDetailed gap analysis:")
for report in quality_reports:
    if report['time_gaps'] > 0:
        print(f"\nAnalyzing gaps for {report['token']}:")
        
        # Reload the file to analyze gaps
        df = pd.read_parquet([pf for pf in sampled_files if report['token'] in pf.name][0])
        df_sorted = df.sort_values('datetime')
        time_diffs = df_sorted['datetime'].diff()
        
        gaps = time_diffs[time_diffs > pd.Timedelta(minutes=1)]
        print(f"Found {len(gaps)} gaps:")
        
        for idx, gap_duration in gaps.items():
            gap_start = df_sorted.loc[idx-1, 'datetime']
            gap_end = df_sorted.loc[idx, 'datetime']
            gap_minutes = gap_duration.total_seconds() / 60
            print(f"  - Gap of {gap_minutes:.0f} minutes: {gap_start} to {gap_end}")
        
        break  # Just show one example

```

```python
# Cell 12: Data Quality Analysis with Polars (more efficient)
print("\n\n=== Data Quality Analysis with Polars ===")

try:
    import polars as pl
    print("Using Polars for efficient data processing")
    use_polars = True
except ImportError:
    print("Polars not installed. Install with: pip install polars")
    print("Continuing with pandas...")
    use_polars = False

if use_polars:
    # Polars version - much faster for multiple files
    quality_reports_pl = []
    
    for pf in sampled_files[:5]:  # Test with 5 files first
        # Read with polars
        df = pl.read_parquet(pf)
        token_name = pf.name.split('_')[0]
        
        # Sort by datetime once
        df = df.sort('datetime')
        
        # Basic stats
        total_rows = len(df)
        date_range = df['datetime'].max() - df['datetime'].min()
        expected_minutes = int(date_range.total_seconds() / 60) + 1
        
        # Check for duplicates
        duplicate_times = total_rows - df['datetime'].n_unique()
        
        # Check for nulls
        null_counts = df.null_count()
        total_nulls = null_counts.sum_horizontal()[0]
        
        # Time gaps analysis
        df = df.with_columns([
            (pl.col('datetime').diff()).alias('time_diff')
        ])
        
        # Count gaps > 1 minute
        gaps = df.filter(pl.col('time_diff') > pl.duration(minutes=1))
        
        # Price quality checks
        zero_prices = df.filter(pl.col('price') <= 0).height
        
        # Price changes
        df = df.with_columns([
            (pl.col('price').pct_change()).alias('price_change')
        ])
        
        extreme_changes = df.filter(pl.col('price_change').abs() > 10).height
        
        quality_reports_pl.append({
            'token': token_name,
            'total_rows': total_rows,
            'expected_rows': expected_minutes,
            'missing_rows': max(0, expected_minutes - total_rows),
            'duplicate_times': duplicate_times,
            'null_values': total_nulls,
            'time_gaps': gaps.height,
            'zero_prices': zero_prices,
            'extreme_changes': extreme_changes,
            'largest_gap_minutes': gaps['time_diff'].max().total_seconds() / 60 if gaps.height > 0 else 0
        })
    
    # Display results
    print("\nPolars Analysis Results:")
    print(f"{'Token':<15} {'Rows':<8} {'Missing':<8} {'Gaps':<6} {'MaxGap':<8} {'Extreme':<8}")
    print("-" * 65)
    
    for report in quality_reports_pl:
        print(f"{report['token']:<15} {report['total_rows']:<8} "
              f"{report['missing_rows']:<8} {report['time_gaps']:<6} "
              f"{report['largest_gap_minutes']:<8.0f} {report['extreme_changes']:<8}")
print("-" * 75)

for report in quality_reports:
    print(f"{report['token']:<15} {report['total_rows']:<8} {report['expected_rows']:<8} "
          f"{report['missing_rows']:<8} {report['time_gaps']:<6} {report['duplicate_times']:<6} "
          f"{report['zero_prices']:<6} {report['extreme_changes']:<8}")

# Detailed analysis of gaps for first token with issues
print("\n\nDetailed gap analysis:")
for report in quality_reports:
    if report['time_gaps'] > 0:
        print(f"\nAnalyzing gaps for {report['token']}:")
        
        # Reload the file to analyze gaps
        df = pd.read_parquet([pf for pf in sampled_files if report['token'] in pf.name][0])
        df_sorted = df.sort_values('datetime')
        time_diffs = df_sorted['datetime'].diff()
        
        gaps = time_diffs[time_diffs > pd.Timedelta(minutes=1)]
        print(f"Found {len(gaps)} gaps:")
        
        for idx, gap_duration in gaps.items():
            gap_start = df_sorted.loc[idx-1, 'datetime']
            gap_end = df_sorted.loc[idx, 'datetime']
            gap_minutes = gap_duration.total_seconds() / 60
            print(f"  - Gap of {gap_minutes:.0f} minutes: {gap_start} to {gap_end}")
        
        break  # Just show one example
```

```python
# Cell 18: Feature Engineering for Prediction
print("\n\n=== Feature Engineering for Memecoin Success Prediction ===")

# Let's analyze the first hour patterns more closely
print("Analyzing first-hour patterns for prediction...")

feature_data = []

for i, pf in enumerate(tokens_with_metadata[:5]):  # Start with 5 tokens
    df = pd.read_parquet(pf)
    address = pf.name.split('_')[-1].replace('.parquet', '')
    metadata = token_map[address]
    
    # First hour analysis
    first_60_min = df.iloc[:60] if len(df) >= 60 else df
    
    # Calculate features
    features = {
        'symbol': metadata['symbol'],
        # Price movement features
        'gain_1h': (first_60_min['price'].iloc[-1] / first_60_min['price'].iloc[0] - 1) * 100,
        'max_gain_1h': (first_60_min['price'].max() / first_60_min['price'].iloc[0] - 1) * 100,
        'volatility_1h': first_60_min['price'].pct_change().std(),
        
        # Price stability features
        'price_changes_1h': (first_60_min['price'].diff() != 0).sum(),  # Number of price changes
        'consecutive_gains': 0,  # Will calculate below
        'largest_drop_1h': 0,  # Will calculate below
        
        # Pattern features
        'time_to_first_2x': None,  # Minutes to first doubling
        'pullback_after_peak': 0,  # How much it pulled back from first peak
        
        # Success metrics (what we're trying to predict)
        'final_gain_24h': (df['price'].iloc[-1] / df['price'].iloc[0] - 1) * 100,
        'volume_24h': metadata['v24hUSD'],
        'success': metadata['v24hUSD'] > 1000000  # Binary success metric
    }
    
    # Calculate consecutive gains
    price_changes = first_60_min['price'].diff()
    gains = (price_changes > 0).astype(int)
    features['consecutive_gains'] = gains.groupby((gains != gains.shift()).cumsum()).sum().max()
    
    # Calculate largest drop
    if len(first_60_min) > 1:
        cummax = first_60_min['price'].cummax()
        drawdown = (first_60_min['price'] - cummax) / cummax
        features['largest_drop_1h'] = drawdown.min() * 100
    
    # Time to first 2x
    double_price = first_60_min['price'].iloc[0] * 2
    doubled = first_60_min[first_60_min['price'] >= double_price]
    if len(doubled) > 0:
        features['time_to_first_2x'] = doubled.index[0]
    
    # Pullback from peak
    peak_1h = first_60_min['price'].max()
    final_1h = first_60_min['price'].iloc[-1]
    features['pullback_after_peak'] = (peak_1h - final_1h) / peak_1h * 100 if peak_1h > 0 else 0
    
    feature_data.append(features)

# Create feature DataFrame
feature_df = pd.DataFrame(feature_data)

print("\nFirst Hour Features vs Success:")
print(feature_df[['symbol', 'gain_1h', 'volatility_1h', 'consecutive_gains', 'time_to_first_2x', 'success']])

# Analyze which features correlate with success
print("\n\nFeature Importance (correlation with 24h volume):")
numeric_features = ['gain_1h', 'max_gain_1h', 'volatility_1h', 'price_changes_1h', 
                   'consecutive_gains', 'largest_drop_1h', 'pullback_after_peak']

for feature in numeric_features:
    corr = feature_df[feature].corr(feature_df['volume_24h'])
    print(f"{feature}: {corr:.3f}")

# Pattern identification
print("\n\nSuccessful Token Patterns (>$1M volume):")
successful = feature_df[feature_df['success']]
if len(successful) > 0:
    print("Average first-hour metrics for successful tokens:")
    print(successful[numeric_features].mean().round(2))
```

```python
# Cell 17: Price-Only Analysis - Multiple Trading Opportunities
print("=== Price-Only Trading Analysis ===")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Analyze price patterns without using any metadata
price_patterns = []

for i, pf in enumerate(parquet_files[:10]):  # Start with 10 tokens
    df = pd.read_parquet(pf)
    token_name = pf.name.split('_')[0]
    
    print(f"\nAnalyzing {token_name}...")
    
    # Calculate rolling metrics
    df['returns'] = df['price'].pct_change()
    df['rolling_max'] = df['price'].rolling(window=60).max()  # 1-hour rolling max
    df['drawdown'] = (df['price'] - df['rolling_max']) / df['rolling_max']
    
    # Identify potential entry points (dips after initial pump)
    df['price_ma_5'] = df['price'].rolling(5).mean()
    df['price_ma_15'] = df['price'].rolling(15).mean()
    
    # Find all significant pumps (>50% in 30 minutes)
    pumps = []
    for i in range(30, len(df)):
        gain_30m = (df['price'].iloc[i] / df['price'].iloc[i-30] - 1) * 100
        if gain_30m > 50:
            pumps.append({
                'start_idx': i-30,
                'end_idx': i,
                'gain': gain_30m,
                'start_time': df['datetime'].iloc[i-30],
                'end_time': df['datetime'].iloc[i]
            })
    
    # Find trading opportunities (buy the dip strategy)
    trades = []
    for i in range(60, len(df)-60):  # Need 60 min before and after
        # Entry signal: price drops 20%+ from recent high
        if df['drawdown'].iloc[i] < -0.2:
            entry_price = df['price'].iloc[i]
            entry_time = df['datetime'].iloc[i]
            
            # Look for exit in next 60 minutes
            future_prices = df['price'].iloc[i+1:i+61]
            
            # Exit strategies
            max_price = future_prices.max()
            max_gain = (max_price / entry_price - 1) * 100
            
            # Stop loss at -10%
            stop_loss_hit = (future_prices < entry_price * 0.9).any()
            
            # Take profit at +30%
            take_profit_hit = (future_prices > entry_price * 1.3).any()
            
            if take_profit_hit:
                exit_idx = future_prices[future_prices > entry_price * 1.3].index[0]
                exit_price = df['price'].loc[exit_idx]
                profit = 30
            elif stop_loss_hit:
                exit_idx = future_prices[future_prices < entry_price * 0.9].index[0]
                exit_price = df['price'].loc[exit_idx]
                profit = -10
            else:
                exit_price = future_prices.iloc[-1]
                profit = (exit_price / entry_price - 1) * 100
            
            trades.append({
                'entry_time': entry_time,
                'entry_price': entry_price,
                'exit_price': exit_price,
                'profit_pct': profit,
                'max_gain': max_gain
            })
    
    price_patterns.append({
        'token': token_name,
        'total_pumps': len(pumps),
        'largest_pump': max([p['gain'] for p in pumps]) if pumps else 0,
        'num_trades': len(trades),
        'profitable_trades': sum(1 for t in trades if t['profit_pct'] > 0),
        'avg_profit': np.mean([t['profit_pct'] for t in trades]) if trades else 0,
        'total_return': sum([t['profit_pct'] for t in trades]) if trades else 0
    })

# Summary
pattern_df = pd.DataFrame(price_patterns)
print("\n\n=== Trading Opportunities Summary ===")
print(pattern_df)
```

```python
from pathlib import Path
import polars as pl

base = Path("data/features")           # or data/features if you prefer
categories = [
    "normal_behavior_tokens",
    "tokens_with_extremes",
    "dead_tokens",
    "tokens_with_gaps",               # optional
]
horizons = [15, 30, 60, 120, 240, 360, 720]

results = {cat: {h: {'up': 0, 'down': 0} for h in horizons}
           for cat in categories}

for cat in categories:
    for fp in (base / cat).glob("*.parquet"):
        df = pl.read_parquet(fp).select("price")
        if df.height < max(horizons) + 1:
            continue
        prices = df["price"].to_numpy()
        for h in horizons:
            future = prices[h:]
            now    = prices[:-h]
            up     = (future > now).sum()
            down   = len(now) - up
            results[cat][h]["up"]   += int(up)
            results[cat][h]["down"] += int(down)

# Pretty-print
for cat in categories:
    print(f"\n=== {cat} ===")
    for h in horizons:
        up   = results[cat][h]['up']
        down = results[cat][h]['down']
        tot  = up + down
        p_up = up / tot * 100 if tot else 0
        print(f"{h:>4}m : UP {up:>8,} ({p_up:5.1f}%)   DOWN {down:>8,}")
```
